```python
import pandas as pd

df = pd.read_csv("../data/raw/amazon_ecommerce_1M.csv")
print("Shape:", df.shape)
print("\nColumns:", df.columns.tolist())
print("\nDtypes:")
print(df.dtypes)
```

```python
candidates = [
    c
    for c in df.columns
    if any(k in c.lower() for k in ["status", "return", "cancel", "ship", "deliver"])
]
print("Candidates:", candidates)
for col in candidates:
    print(f"\n--- {col} ---")
    print(df[col].value_counts(dropna=False).head(20))
```

```python
print(df.nunique().sort_values(ascending=False))
```

```python
print("=== First 3 rows ===")
print(df.head(3).to_string())
print("\n=== Missing % per column ===")
print((df.isna().mean() * 100).round(2).sort_values(ascending=False).head(20))
```
